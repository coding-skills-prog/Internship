Customer satisfaction has emerged as one of the most important factors that guarantee the success of online store; it has been posited as a key stimulant of purchase, repurchase intentions and customer loyalty. A comprehensive review of the literature, theories and models have been carried out to propose the models for customer activation and customer retention. Five major factors that contributed to the success of an e-commerce store have been identified as: service quality, system quality, information quality, trust and net benefit. The research furthermore investigated the factors that influence the online customers repeat purchase intention. The combination of both utilitarian value and hedonistic values are needed to affect the repeat purchase intention (loyalty) positively


```python
# Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading a dataset

df = pd.read_csv('Customer_retention_dataset.csv', index_col=None, header=0, encoding='latin1')
df.head()
```

```python
df.duplicated().sum() #to check with duplicate records
```

So,there are duplicate records in dataset and we can drop these records from dataset.

```python
#droping the duplicate records
df = df.drop_duplicates()
```

```python
#to check how many rows and columns in dataframe
df.shape
```

```python
#to get the info of this dataframe
df.info()
```

So all columns are objective type except for column "4 What is the Pin Code of where you shop online from?" which is an Int type.

So,there are no null values.

## Exploratory data analysis

```python
import warnings
warnings.filterwarnings('ignore') #to remove warning messages
```

```python
#using a factorplot from seaborn
sns.factorplot('1Gender of respondent', data=df, kind="count") ## Let's check gender
```

the respondent was mostly a female compared to male

```python
#Now lets separate gender with city

sns.factorplot('1Gender of respondent', data=df, kind="count", hue='3 Which city do you shop online from?')
```

### So,with above observation we can say that,
#### -> mostly male person who is shoping for online from Delhi or Noida.
#### -> mostly female person who is shoping for online from Greater Noida,Bangalore or Karnal.

```python
#Now lets separate gender with age

sns.factorplot('1Gender of respondent', data=df, kind="count", hue='2 How old are you? ')
```

#### So,usually the female respondant aged between 21 to 30 or between 41-50 years.
#### and the male respondant aged between 31-40 or between 41-50 years.

```python
#Now lets check how many times customer made a purchase in the past 1 year 

a = sns.factorplot('6 How many times you have made an online purchase in the past 1 year?', data=df, kind="count")
a.set_xticklabels(rotation=70)
```

SO,most of them went for online shoping for less than 10 times in the past 1 year.

```python
#now,lets check that how long since the customer is shopiing for online
a = sns.factorplot('5 Since How Long You are Shopping Online ?', data=df, kind="count")
a.set_xticklabels(rotation=70)
```

SO,most of them are shopping for online products from last 4 years and beyound.

```python
#lets check with how much time will customer consume in searching for a product considering their age

a = sns.factorplot('14 How much time do you explore the e- retail store before making a purchase decision?                                                                   ', data=df, kind="count", hue='2 How old are you? ')
a.set_xticklabels(rotation=70)
```

SO,most customers whose age lies in between 21-30 years goes for searching a product for about more then 15 mins before buying any product.

```python
#checking with leaving behind the added product in cart without buying it considering how long the customer has been shopping for online.
a = sns.factorplot('16 How frequently do you abandon (selecting an items and leaving without making payment) your shopping cart?\t\t\t\t\t\t\t                                                   ', data=df, kind="count", hue='5 Since How Long You are Shopping Online ?')
a.set_xticklabels(rotation=70)
```

So,mostly customer whose been shoping for more then 4 years is more often purchases the item with exception for sometimes the customer abondon the added product in their cart.

```python
#reason to abondon the product in cart.

a = sns.factorplot('17 Why did you abandon the Bag, Shopping Cart?\t\t\t\t\t                                                                                                                                                       ', data=df, kind="count")
a.set_xticklabels(rotation=70)
```

Most of them thought that they had better alternative offer before abondoning the product in cart itself.

```python
#to check with customer feedback on recommended products.

a = sns.factorplot('16 How frequently do you abandon (selecting an items and leaving without making payment) your shopping cart?\t\t\t\t\t\t\t                                                   ', data=df, kind="count", hue='19 Information on similar product to the one highlighted  is important for product comparison')
a.set_xticklabels(rotation=70)
```
