## 1) Write a python program to display all the header tags from wikipedia.org


```python
def web_scarpping_1(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)
  header_tags = []
  for i in soup.find_all('h2',class_='mp-h2'):
    header_tags.append(i.text)
  
  #removing an unwanted substring on second element of list
  header_tags[1] = header_tags[1].replace("\xa0",'')
  print("Header tags ->\n")
  return header_tags

web_scarpping_1("https://en.wikipedia.org/wiki/Main_Page")
```

## 2)Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
import pandas as pd
def web_scarpping_2(url1,url2):
  #requesting for a source code of the web page from the url
  page1 = requests.get(url1)
  soup1 = BeautifulSoup(page1.content)
  page2 = requests.get(url2)
  soup2 = BeautifulSoup(page2.content)

  title,rating_accu,year_of_release,title2,rating_accu2,year_of_release2= [],[],[],[],[],[]

  #to get movie title and year of release,we can use "h3" label and 'lister-item-header' as class
  #to get ratings,we can use "span" label and "ipl-rating-star__rating" as class

  #for page1

  for i in soup1.find_all('h3',class_='lister-item-header'):
    title.append(((i.text).split("\n"))[2])
    year_of_release.append((((i.text).split("\n"))[3]).strip("()"))
  
  for i in soup1.find_all("div",class_="inline-block ratings-imdb-rating"):
    rating_accu.append((i.text).strip("\n"))

  #for page2

  for i in soup2.find_all('h3',class_='lister-item-header'):
    title2.append(((i.text).split("\n"))[2])
    year_of_release2.append((((i.text).split("\n"))[3]).strip("()"))
  
  for i in soup2.find_all("div",class_="inline-block ratings-imdb-rating"):
    rating_accu2.append((i.text).strip("\n"))

  # making of a Dataframe from the extracted data of web page
  data = {'Title of Movie':title,'Rating':rating_accu,"Year of release":year_of_release}
  df = pd.DataFrame(data)


  data2 = {'Title of Movie':title2,'Rating':rating_accu2,"Year of release":year_of_release2}
  df2 = pd.DataFrame(data2)

  #concatanating two dataframes
  df_final = pd.concat([df,df2])
  print(df_final.count())
  print("======================")
  return df_final


web_scarpping_2("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc","https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt") 
```

##3)Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import pandas as pd
def web_scarpping_3(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)

  title,rating_accu,rating,year_of_release= [],[],[],[]

#to get movie title and year of release,we can use "h3" label and 'lister-item-header' as class
#to get ratings,we can use "span" label and "ipl-rating-star__rating" as class

  for i in soup.find_all('h3',class_='lister-item-header'):
    title.append(((i.text).split("\n"))[2])
    year_of_release.append((((i.text).split("\n"))[3]).strip("()"))
    

  for j in soup.find_all("span",class_="ipl-rating-star__rating"):
    rating_accu.append(j.text)


  for k in range(0,len(rating_accu),23):
    rating.append(rating_accu[k])


#making of a Dataframe from the extracted data of web page
  data = {'Title of Movie':title,'Rating':rating,"Year of release":year_of_release}
  df = pd.DataFrame(data)
  return df

  

web_scarpping_3("https://www.imdb.com/list/ls009997493/?sort=user_rating,desc&st_dt=&mode=detail&page=1") 
```

##4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 





```python
import pandas as pd
def web_scarpping_4(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)

  product_name,price_tag,discount= [],[],[]

  #extracting a product name
  for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    product_name.append(i.text)
  
  #extracting a price of product
  for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    price_tag.append((i.text).strip("₹"))

  #extracting a discount of product
  for i,j in zip(soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"),soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm")):
    discount.append(("(" + (i.text).strip("off ") + ")") + ((j.text).strip("₹")).split()[0])
 
  #Creating a dataframe using an extracted data
  data = {'Product name':product_name,'Price':price_tag,"Discount":discount}
  df = pd.DataFrame(data)
  return df
web_scarpping_4("https://meesho.com/bags-ladies/pl/p7vbp")
```

## 5) Write a python program to scrape cricket rankings from icc-cricket.com.

###a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def web_scarpping_5a(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)

  teams,mat,points,rating = [],[],[],[]

  #extracting a first team's detail seperatly which is having a different class compared to other teams' class.
  for i in soup.find_all('tr',class_="rankings-block__banner"):
    if len(teams) < 10:
      teams.append(((i.text).split("\n"))[4])
      mat.append(((i.text).split("\n"))[7])
      points.append(((i.text).split("\n"))[8])
      rating.append(((i.text).split("\n"))[10])
    else:
      break

  #extracting a next 9 teams' details.
  for j in soup.find_all('tr',class_="table-body"):
    if len(teams) < 10:
      teams.append(((j.text).split("\n"))[4])
      mat.append(((j.text).split("\n"))[7])
      points.append(((j.text).split("\n"))[8])
      rating.append(((j.text).split("\n"))[9])
    else:
      break

  #Creating a dataframe using an extracted data
  data = {'Team':teams,'Matches':mat,"Points":points,"Rating":rating}
  df = pd.DataFrame(data)
  print("         Men's ODI Team Rankings(Top 10)\n")
  return df
web_scarpping_5a("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

## 7)Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.


```python

def web_scarpping_7(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)

  title,date,content,vid_link= [],[],[],[]

  for i in soup.find_all('h2',class_="entry-title"):
    title.append(i.text)
  for i in soup.find_all('p',class_="entry-meta"):
    if i.text[:5] != "Filed":
      date.append(((i.text).split("by"))[0])
  for i in soup.find_all('div',class_="entry-content"):
    if "https" in i.text:
      content.append((i.text.split("\nVSCode"))[0].strip("\n"))
    else:
      content.append(i.text.strip("\n"))
  for link in soup.find_all('iframe',class_="youtube-player"):
    vid_link.append(link["src"])
  del title[4]
  del content[4]
  del date[4]

  # Creating a dataframe using an extracted data
  data = {'Heading':title,'Date':date,"content":content,"Video link":vid_link}
  df = pd.DataFrame(data)
  return df
  return title

web_scarpping_7("https://coreyms.com/")
```

##9) Write a python program to scrape mentioned details from dineout.co.in :
###i) Restaurant name
###ii) Cuisine
###iii) Location
###iv) Ratings
###v) Image URL 

```python
def web_scarpping_9(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)

  Restaurant,location,cuisine,rating,Image_URl = [],[],[],[],[]
  for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Restaurant.append(i.text)
  for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text) 
  for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(((i.text).split("|"))[1])
  for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text) 
  for i in soup.find_all('img',class_="no-img"):
    Image_URl.append(i["data-src"]) 

  #Creating a dataframe using an extracted data
  data = {'Restaurant name':Restaurant,'Cuisine':cuisine,'Location':location,'Rating':rating,'Image_URl':Image_URl}
  df = pd.DataFrame(data)
  return df


web_scarpping_9("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

##10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts . 

```python
def web_scarpping_10(url):
  #requesting for a source code of the web page from the url
  page = requests.get(url)
  soup = BeautifulSoup(page.content)

  name,price,Img_URL = [],[],[]
  for i,j,k in zip(soup.find_all('h3'),soup.find_all("span",class_="discountedPriceText"),soup.find_all('img')):
    if len(name) != 10:
      name.append(i.text)
      price.append(j.text)
      Img_URL.append(k["src"])
    else:
      break
  #Creating a dataframe using an extracted data
  data = {'Product name':name,'Price':price,'Image URL':Img_URL}
  df = pd.DataFrame(data)
  return df
web_scarpping_10("https://www.bewakoof.com/women-tshirts?ga_q=tshirts")
```
